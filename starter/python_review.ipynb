# Matrices and Nested Lists

```python
# In Python, matrices are often represented as lists of lists.
matrix = [
    [1, 2, 3],
    [4, 5, 6],
    [7, 8, 9]
]

print(matrix)
print(matrix[0])
print(matrix[1][2])  # row 2, column 3
```

### Looping Through a Matrix

```python
for row in matrix:
    for value in row:
        print(value, end=" ")
    print()
```

# Conditional Statements (if / elif / else)

```python
score = 85

if score >= 90:
    grade = "A"
elif score >= 80:
    grade = "B"
else:
    grade = "C"

print("Grade:", grade)
```

# Classes and Objects

```python
# Classes allow you to model real-world objects.
class Student:
    def __init__(self, name, course):
        self.name = name
        self.course = course

    def introduce(self):
        return f"My name is {self.name} and I am enrolled in {self.course}"
```

```python
student1 = Student("Jessica", "AIG 230")
print(student1.introduce())
```

# Putting It All Together

```python
# TODO
""" Create a class called courses_winter 2026
with attributes course_name, code, time, grade and a method to display course info."""

class course_winter_2026:
  def __init__(self, course_name, code, time, grade):
    self.course_name = course_name
    self.code = code
    self.time = time
    self.grade = grade

  def Display(self):
    return f"This couse is called {self.course_name}{self.code} and it starts at {self.time}. Your current grade is {self.grade}"
```

```python
# TODO
""" Create three instances of the courses_winter 2026 class and display their info."""
course1= course_winter_2026("NLP ", "AIG230", "12:45 AM", "A")
course2= course_winter_2026("DL ", "AIG220", "10:45 AM", "C")
course3= course_winter_2026("CV ", "AIG210", "9:00 AM", "B")

print(course1.Display())
print(course2.Display())
print(course3.Display())

```
